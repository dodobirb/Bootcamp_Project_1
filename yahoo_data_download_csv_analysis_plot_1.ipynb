```python
import yfinance as yf
import pandas as pd
from pathlib import Path
```

```python
data = yf.download("JPM BAC C WFC GS MS PNC USB COF BNY STT KEY FITB TD", start="2008-01-31", end="2021-04-01", interval="1mo")

stock_data_df = pd.DataFrame(data).dropna()

stock_data_df.drop(columns=["Volume","Open","Low","High","Close"],axis=1,level=0,inplace=True)

display(stock_data_df.shape)
display(stock_data_df)
```

```python
stock_data_df.to_csv("stockdata_2.csv")
```
